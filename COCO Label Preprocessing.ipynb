```python
def normalize_bbox(bbox_list, image_w, image_h):
    """
        COCO dataset bbox format is (top_left_x, top_left_y, width, height)
        YOLO requires it in (x_center, y_center, width, height) format normalized between 0 and 1 
    """
    
    x1, y1, w, h = bbox_list
    
    return [((2*x1 + w)/(2*image_w)) , ((2*y1 + h)/(2*image_h)), w/image_w, h/image_h]    
```

```python
def pad_image_id(image_id):
    """
    Image names are padded with 0s at the start, but image_ids are not
    YOLO requires image name and label name to be same
    """
    
    number_of_zeros = 12 - len(str(image_id))
    
    padded_image_id = (number_of_zeros*'0') + str(image_id)
    
    return padded_image_id
```

```python
for items in train_labels_mod['annotations']:
    
    with open(train_dir+"labels//"+ pad_image_id(items["image_id"]) + '.txt','a') as f:
        
        f.write(str(items['category_id'])+" ")
        
        img_height = None
        img_width = None
        
        for image_info in train_labels_mod['images']:
            
            if image_info['id'] == items['image_id']:
                
                img_height = image_info['height']
                
                img_width = image_info['width']
            
        bbox = normalize_bbox(items['bbox'], img_width, img_height)
        
        for items in bbox:
            f.write(str(items)+" ")
        
        f.write('\n')
```
